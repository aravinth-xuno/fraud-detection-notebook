```python
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
model = IsolationForest(contamination=0.001)
```

```python
train_df = pd.read_csv("/content/drive/Shareddrives/transaction-training-set/dataset/transformed_dataset_high_velocity_users_features.csv", index=False)
```

```python
features = [
    "last_1hour_count",
    "last_24hour_count",
    "last_30days_count",
    "ratio_1hr_24hr",
    "ratio_24hr_30day"
]
```

```python
model.fit(train_df[features])
```

```python
import pickle
with open("/content/drive/Shareddrives/transaction-training-set/model/v0.01_high_velocity_model", "wb") as file:
    pickle.dump(model, file)
    print("Model saved Successfully")
```
